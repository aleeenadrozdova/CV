### Задание 1.
Даны комплексные числа
$$
z_1 = 5\sqrt{3} + 5i,\quad z_2 = -3 + 3i,\quad z_3 = - 1 + i,\quad z_4 = i,\quad z_5 = \frac{12}{13} + \frac{5}{13}i,
$$
вычислить (представить в алгебраической форме) с помощью sympy
$$
z_1 \cdot z_2,\quad (z_2 - 5z_3)^6,\quad \sqrt{\frac{z_4}{z_3}},\quad \frac{z_1^2 + z_3}{13z_5 - 2z_2}.
$$

```python
z1 = 5*3**(sympy.S.Half) + 5*I
z2 = -3 + 3*I
z3 = -1 + I
z4 = I
z5 = sympy.S(12)/13 + sympy.S(5*I)/13

display(sympy.expand_complex(z1*z2))
display(sympy.expand_complex((z2 - 5*z3)**6))
display(sympy.expand_complex((z4/z3)**(sympy.S.Half)))
display(sympy.expand_complex((z1**2 + z3)/(13*z5 - 2*z2)))
```

### Задание  3.
Изобразить на координатной плоскости точки, соответствующие корням  $7$-й степени комплексного числа $z = \frac{3}{5} + i\frac{3}{5}$. Подписать точки $A_k$, где $k$  - номер корня.

```python
n = 7
z = sympy.S(3)/5 + sympy.S(3)/5*I
roots_list_7 = [sympy.root(z, n, k) for k in range(n)]
roots_7 = [sympy.expand_complex(item) for item in roots_list_7]
display(*roots_7)
roots_x = [sympy.re(item) for item in roots_list_7]
roots_y = [sympy.im(item) for item in roots_list_7]
plt.scatter(roots_x, roots_y)
for k, x_coord in enumerate(roots_x): 
    plt.annotate('$A_' + str(k) + '$',
            xy=(x_coord, roots_y[k]), 
            xytext=(x_coord + 0.05, roots_y[k] - 0.05)) 
plt.axis('equal')
plt.xlabel("Re")
plt.ylabel("Im")
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
```

### Задание  4.
Изобразить на координатной плоскости точки, соответствующие $z_4 = - \sqrt{3} + 3i$ и $\bar{z_4}$, подписать $z_4$ тригонометрической формой, $\bar{z_4}$ - экспоненциальной ($z = r e^{i\varphi}$, $r$ и $\varphi$ такие же, как в тригонометрической форме).

```python
z4 = -3**(sympy.S.Half) + 3*I
r4 = sympy.functions.Abs(z4)
phi4 = sympy.functions.arg(z4)
z4_ = sympy.conjugate(z4)
r4_ = sympy.functions.Abs(z4_)
phi4_ = sympy.functions.arg(z4_)
roots_x = [sympy.re(z4), sympy.re(z4_)]
roots_y = [sympy.im(z4), sympy.im(z4_)]
plt.scatter(roots_x, roots_y)
plt.annotate(fr'$z_4 = {latex(r4)} \left(\cos \left({latex(phi4)}\right) + i \sin \left({latex(phi4)}\right)\right)$',
            xy=(roots_x[0], roots_y[0]), 
            xytext=(roots_x[0] + 0.25, roots_y[0] - 0.05)) 
plt.annotate(fr'$z_4 = {latex(r4_)} e^{{i\left({latex(phi4_)}\right)}}$',
            xy=(roots_x[1], roots_y[1]), 
            xytext=(roots_x[1] + 0.25, roots_y[1] - 0.05)) 
plt.axis('equal')
plt.xlabel("Re")
plt.ylabel("Im")
plt.xlim(-4.5, 4.5)
plt.ylim(-4, 4)
```

Вывести алгебраическую форму полученных корней на экран и сравнить с результатом, полученным с помощью sympy.root (с использованием sympy.expand_complex для упрощения выражения). Аналитическое представление оказывается разным.

Корни уравнения вывести на экран в алгебраической, тригонометрической и экспоненциальной форме.

$(1 + 9 i)x^2 + (84 - 64 i)x  -224 - 48 i = 0$

```python
x = sympy.Symbol('x')
f = (1 + 9*I)*x**2 + (84 - 64*I)*x  -224 - 48*I
roots = [*sympy.roots(f)]
for k, root in enumerate(roots):
    phi = sympy.simplify(sympy.functions.arg(root))
    r = sympy.functions.Abs(root)
    display(Latex(fr'z_{k+1} = {sympy.latex(root)} = {latex(r)} \left(\cos \left({latex(phi)}\right) + i \sin \left({latex(phi)}\right)\right) = {latex(r)} e^{{i\left({latex(phi)}\right)}}'))
```
