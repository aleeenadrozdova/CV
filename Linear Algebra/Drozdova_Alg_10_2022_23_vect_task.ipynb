### Задание 1.
Задать систему координат с именем '1'. 
Найти скалярное и векторное произведение векторов $a(-\cos(\pi/4), \sin(\pi/3), 1)$ и $b(2, -1, 3)$.

```python
from sympy.vector import CoordSys3D
N = CoordSys3D('1')
display(N)
a = -Cos(Pi/4)*N.i + Sin(Pi/3)*N.j + N.k
b = 2*N.i - 1*N.j + 3*N.k
display(Latex("""a\\cdot b = {0},\\\ a\\times b = {1}""".format(*map(latex, (a.dot(b), a.cross(b))))))
```

### Задание 2
Ввести безымянную систему координат M и символы $p$ и $q$.
Вычислить $((v + u)\cdot(v - u)) + (((2v + u)\times(v - 3u))\cdot(2v - 3u)) - ((3v - 2u)\cdot((5v - u)\times(- u)))$, $u(q, -p, q + p)$, $v(p, 2p - q, 1)$. Скалярное произведение обозначено $\cdot$, векторное $\times$.
Использовать  & и ^, скобки расставить как в условии, поскольку приоритет у & и ^ ниже, чем у + и -. Упростить полученное выражение.

```python
M = CoordSys3D('')
p, q = symbols('p q')
u = q*M.i - p*M.j + (q+p)*M.k
v = p*M.i + (2*p - q) * M.j + M.k
expr = ((v + u)&(v - u)) + (((2*v + u)^(v - 3*u))&(2*v - 3*u)) - ((3*v - 2*u)&((5*v - u)^(- u)))
display(Latex(f"$((v + u)\\cdot(v - u)) + (((2v + u)\\times(v - 3u))\\cdot(2v - 3u)) - ((3v - 2u)\\cdot((5v - u)\\times(- u))) = {latex(sympy.simplify(expr))}$"))
```

### Задание 3
Упростить вектор $g(-2p^2 + 5pq - 3q^2, \cos(q) + 2\sin(q - \pi/2), \sin^2(p) - \cos^2(p))$.
Вывести на экран координаты упрощенного вектора, его длину и проекции на координатные оси

```python
g = (-2*p**2 + 5*p*q - 3*q**2)*M.i + (Cos(q) + 2*Sin(q - Pi/2))*M.j + (Sin(p)**2 - Cos(p)**2)*M.k
display(Latex('g = ' + ' = '.join(map(latex, (g, g.factor().trigsimp())))))
g = g.factor().trigsimp()
display(*map(Latex, map('\ =\ '.join,
                        zip(('coordinates', 'length', 'projection_i', 'projection_j', 'projection_k'),
                            map(latex,
                                (g.components, g.magnitude(), M.i.projection(g), M.j.projection(g), M.k.projection(g)))))))
```

### Задание 5.
Ввести новую систему координат S5, которая получается поворотом системы координат M из Задания 2 на угол $\pi/4$ относительно оси $j$ по часовой стрелке.
Определить вектор b5 на основе матрицы из Задания 4 в соответствии с системой координат M, затем получить представление в матричном виде вектора b5_S5 в системе координат S5.  Вывести на экран матрицу поворота, вектор b5 в системе координат M  и S5.

```python
S5 = M.orient_new_axis('S5new', -sympy.pi / 4, M.j)
b5_M = matrix_to_vector(b, M)
b5_S5 = b5_M.to_matrix(S5)
R_matr = M.rotation_matrix(S5)
display(*map(Latex, map('\ =\ '.join,
                        zip(('R\_matr', 'b5\_M', 'b5\_S5', 'sympy.simplify(R\_matr*b5\_Sys5\_new)'),
                            map(latex,
                                (R_matr,
                                 b5_M,
                                 sympy.simplify(sympy.expand(b5_S5))))))))
```

### Задание 6.
Повернуть систему координат  M  из Задания 2 на $\pi/4$ по часовой стрелке относительно оси, определяемой вектором $(1, -1, 0)$. Выразить вектор b5 в новой системе координат.

```python
S6 = M.orient_new_axis('S6', -sympy.pi / 4, M.i - M.j)
b5_S6 = sympy.simplify(sympy.expand(express(b5_M, S6)))
b5_S6_matrix = b5_M.to_matrix(S6)
display(Latex(',\\\ '.join(map('\ =\ '.join,
                        zip(('b5\_S6', 'b5\_S6\_matrix'),
                            map(latex,
                                (b5_S6, sympy.simplify(sympy.expand(b5_S6_matrix)))))))))
```

### Индивидуальное задание.
Ввести безымянную систему координат XYZ, задать в ней векторы $v_1(-9, -7, 2)$ и $v_2(7, 9, 3)$. Построить новую систему координат XYZ1 с именем R поворотом  XYZ на $\pi/3$ против часовой стрелки относительно оси, определяемой вектором $(6, -9, 0)$, выразить векторы $v_1$ и $v_2$ в новой системе координат. Найти скалярное и векторное произведение векторов в новой и старой системе координат. Выразить вектор, равный векторному произведению $v_1$ и $v_2$ в старой системе координат через новую систему координат, сравнить с векторным произведением $v_1$ и $v_2$, выраженных в новой системе координат.


```python
XYZ = CoordSys3D('')
v1 = -9*XYZ.i - 7*XYZ.j + 2*XYZ.k
v2 = 7*XYZ.i + 9*XYZ.j +3*XYZ.k
XYZ1 = XYZ.orient_new_axis('R', sympy.pi / 3, 6*XYZ.i - 9*XYZ.j)
v1_R = sympy.simplify(sympy.expand(express(v1, XYZ1)))
v2_R = sympy.simplify(sympy.expand(express(v2, XYZ1)))
display(*map(Latex, map('\ =\ '.join,
                        zip(('v1', 'v2', 'v1_R', 'v2_R', 'v1 \\cdot v2', 'v1 \\times v2', 'v2 \\times v1',
                             'v1_R \\cdot v2_R', 'v1_R \\times v2_R', 'v2_R \\times v1_R', '(v1 \\times v2)_R'),
                            map(latex,
                                (v1, v2, v1_R, v2_R, v1.dot(v2), v1.cross(v2), v2.cross(v1),
                                 sympy.simplify(v1_R.dot(v2_R)), sympy.simplify(v1_R.cross(v2_R)), sympy.simplify(v2_R.cross(v1_R)),
                                 express(v1.cross(v2), XYZ1)
                                 ))))))
```
