#### –£—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤ —Ñ–æ—Ä–º–µ $f(x)=0$

```python
import numpy as np
import sympy
from sympy import nsolve
import matplotlib.pyplot as plt
%matplotlib inline
```

### –ó–∞–¥–∞—á–∞ 1
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $ùë•^4‚àíe^x=0$ –ø—Ä–∏ $-2 \le x \le 2$, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è —Å –æ—Å—å—é –û–•.

```python
def f(x):
    return x**4-np.e**x
x = sympy.Symbol("x")
X = np.linspace(-2, 2, 100)
plt.plot(X, f(X))
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
roots = [sympy.nsolve(f(x), x, point) for point in [-1, 1]]
roots
```

```python
ax = plt.gca() 
for i in range(len(roots)): 
    ax.annotate('A' + str(i + 1),
            xy=(roots[i], 0), xycoords='data',
            xytext=(roots[i]-0.1, 1), textcoords='data')
X = np.linspace(-2, 2, 100)
Y = f(X)
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.plot(X, Y, color='green', label=r'$f(x) = x^4-e^x$')
ax.legend(loc='upper right')
ax.scatter(roots, [0 for x in range(len(roots))])
```

### –ó–∞–¥–∞—á–∞ 2.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x)=\ln x$ –ø—Ä–∏ $0< x \le 3$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
def f(x, lib='sympy'):
    if lib == 'sympy':
        return (sympy.sin(2*x))**2
    if lib == 'numpy':
        return (np.sin(2*x))**2
    return 'error'
def g(x, lib='sympy'):
    if lib == 'sympy':
        return sympy.log(x)
    if lib == 'numpy':
        return np.log(x)
    return 'error'
x = sympy.Symbol("x")
X = np.linspace(0.0001, 3, 100)
plt.plot(X, f(X, 'numpy') - g(X, 'numpy'))
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
roots = [sympy.nsolve(sympy.Eq(f(x), g(x)), x, point) for point in [1, 2, 2.5]]
roots
```

```python
y_coords = [f(x) for x in roots]
ax = plt.gca() 
for i in range(len(roots)): 
    ax.annotate('A' + str(i + 1),
            xy=(roots[i], y_coords[i]), xycoords='data',
            xytext=(roots[i]-0.05, y_coords[i]+0.2), textcoords='data')
X = np.linspace(0, 3, 100)
Y_f = f(X, 'numpy')
Y_g = g(X, 'numpy')
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.plot(X, Y_f, color='green', label=r'$f(x) = \sin^2(2x)$')
ax.plot(X, Y_g, color='skyblue', label=r'$g(x) = \ln x$')
ax.legend(loc='lower right')
ax.scatter(roots, y_coords)
```

### –ó–∞–¥–∞—á–∞ 3.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x^2-7x+6)=3/4$ –Ω–∞ $[3, 4]$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
def f(x, lib='sympy'):
    if lib == 'sympy':
        return (sympy.sin(2*x**2-7*x))**2
    if lib == 'numpy':
        return (np.sin(2*x**2-7*x))**2
    return 'error'
x = sympy.Symbol("x")
X = np.linspace(3, 4, 100)
plt.plot(X, f(X, 'numpy')-0.75)
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 3))
ax.spines['bottom'].set_position(('data', 0))
roots = [sympy.nsolve(f(x, 'sympy')-0.75, x, point) for point in [3, 3.2, 3.65, 3.7]]
roots
```

```python
y_coords = [f(x) for x in roots]
ax = plt.gca() 
for i in range(len(roots)): 
    ax.annotate('A' + str(i + 1),
            xy=(roots[i], y_coords[i]), xycoords='data',
            xytext=(roots[i]-0.01, y_coords[i]+0.1), textcoords='data')
X = np.linspace(3, 4, 100)
Y_f = f(X, 'numpy')
Y_g = [0.75 for x in range(len(X))]
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 3))
ax.spines['bottom'].set_position(('data', 0))
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(3, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.plot(X, Y_f, color='green', label=r'$f(x) = \sin^2(2x^2-7x+6)$')
ax.plot(X, Y_g, color='skyblue', label=r'$g(x) = \frac{3}{4}$')
ax.legend(loc='lower right')
ax.scatter(roots, y_coords)
```

### –ó–∞–¥–∞—á–∞ 4.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ ${\rm tg}^4(2x^2+1)-10=0$ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $[0, 0.2]$ –º–µ—Ç–æ–¥–æ–º –¥–∏—Ö–æ—Ç–æ–º–∏–∏. –ì—Ä–∞—Ñ–∏–∫ —Å—Ç—Ä–æ–∏—Ç—å –Ω–µ –Ω—É–∂–Ω–æ.

```python
def f(x):
    return (sympy.tan(2*x**2 + 1))**4 - 10
root = sympy.nsolve(f(x), x, [0, 0.2], solver='bisect', verify=False, prec=10)
print(root)
print(f(root)) #–ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –Ω–æ–ª—å
```

### –ó–∞–¥–∞—á–∞ 5.
–†–µ—à–∏—Ç—å —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π 
$$
\left\{\begin{matrix}
(x - 1)^2 + (y - 2)^2 + (z + 3)^2 = 25\\
x + y + z = 0\\
-2x + 3y - z = 7
\end{matrix}\right.
$$
–Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ (0, 0, 0).
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–æ–π.

```python
x, y, z = sympy.symbols('x y z')
System = [sympy.Eq((x-1)**2+(y-2)**2+(z+3)**2, 25), sympy.Eq(x+y+z, 0), sympy.Eq(-2*x+3*y-z, 7)]
root = sympy.nsolve(System, [x, y, z], [0, 0, 0], prec=20)
display(root)
display(*[j.subs([(x, root[0]), (y, root[1]), (z, root[2])]) for j in System])
```

### –ò–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ
–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π $f(x)$ –∏ $g(x)$, —É—Ç–æ—á–Ω–∏—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–µ–∫ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è, —Ä–µ—à–∞—è —á–∏—Å–ª–µ–Ω–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ. –ù–∞  –≥—Ä–∞—Ñ–∏–∫–µ –æ—Ç–º–µ—Ç–∏—Ç—å –∏ –ø–æ–¥–ø–∏—Å–∞—Ç—å –±—É–∫–≤–∞–º–∏ $–ê_1$, $A_2$, ... —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

$f(x)=-\frac{x^2}{4}+x+10$


$g(x) = 2x^2 + 10x + 10$

```python
def f(x):
    return -x**2/4+x+10
def g(x):
    return 2*x**2+10*x+10
x = sympy.Symbol("x")
ax = plt.gca()  
X = np.linspace(-8,2, 100)
Y_f = f(X)
Y_g = g(X)
ax.plot(X, Y_f, color='green', label="$f(x) = -\\frac{x^2}{4}+x+10$")
ax.plot(X, Y_g, color='skyblue', label='$f(x) = 2x^2+10x+10$')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
roots = sympy.solve(sympy.Eq(f(x), g(x)), x)
y_coords = [f(x) for x in roots] 
ax.legend(loc='upper left')
for i in range(len(roots)): 
    ax.annotate('A' + str(i + 1),
            xy=(roots[i], y_coords[i]), xycoords='data',
            xytext=(roots[i]-0.2, y_coords[i]+3), textcoords='data')
ax.scatter(roots, y_coords)
```
