### Задание 1
Вычислить площадь фигуры, ограниченной графиками функций $f = 2 + x^2$ и  $g = 2 - x^3$. Построить графики, найти точки пересечения, затем вычислить соответствующие интегралы.

```python
x = Symbol('x')
f = 2 + x**2
g = 2 - x**3
plot(f, g, (x, -1.1, 0.1), legend = True)
roots = solveset(sympy.Eq(f, g), x, domain=S.Reals)
S_fg = sympy.simplify((f - g).integrate((x, min(roots), max(roots))))
display(roots, S_fg)
```

### Задача 2.
Вычислить площадь фигуры, ограниченной графиком функции $x = \cos^3(t)$, $y = \sin^3(t)$, $0 \le t \le 2\pi$.

```python
t = Symbol('t')
f = Cos(t)**3
g = Sin(t)**3
roots = solveset(sympy.Eq(f, g), domain=Interval.Ropen(0, 2*Pi))
S_fg = sympy.simplify((f - g).integrate((t, min(roots), max(roots))))
display(roots, S_fg)
```
